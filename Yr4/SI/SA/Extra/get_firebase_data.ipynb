```python
#['action','id', 'day','timestamp', 'valueX', 'valueY', 'valueZ', 'magnitude']

def fall_detection(records):

    detection_counter = 0

    for record in records:
        detected = False
        yValues = []
        magnitude_values = []
        timestamp_values = []

        for timestamp, details in record.items():

            magnitude = math.sqrt(details['valueX']**2 + details['valueY']**2 + details['valueZ']**2)
            magnitude_values.append(magnitude)
            timestamp_values.append(timestamp)
            yValues.append(details['valueY'])

            if magnitude >= 20:
                detected = True

        print(int(timestamp_values[len(timestamp_values)-1]) - int(timestamp_values[0]))

        if detected:
            max_magnitude_index = magnitude_values.index(max(magnitude_values))
            print(max_magnitude_index)
            start_index = max_magnitude_index

            while start_index > 0 and (int(timestamp_values[max_magnitude_index]) - int(timestamp_values[start_index])) < 2000:
                start_index -= 1

            end_index = max_magnitude_index
            while end_index < len(timestamp_values)-1 and (int(timestamp_values[end_index]) - int(timestamp_values[max_magnitude_index])) < 2000:
                end_index += 1

            mean_magnitude = sum(magnitude_values[start_index:end_index + 1]) / (end_index - start_index + 1) 
            after_fall = max(magnitude_values[end_index:]) - min(magnitude_values[end_index:])
   
            
            if mean_magnitude >= 9.5:
                print(datetime.fromtimestamp(int(timestamp_values[end_index]) / 1000.0))
                print(max(magnitude_values[end_index:]), min(magnitude_values[end_index:]))
                print(end_index, len(magnitude_values))
                detection_counter += 1

    return len(records), detection_counter


```
